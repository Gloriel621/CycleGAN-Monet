```python
# CycleGAN architecture and trainer available at
# https://github.com/Gloriel621/CycleGAN-Monet

import torch
from torchvision import transforms

from dataset import ImageDataset
from trainer import CycleGAN_Trainer
```

```python
# Hyperparameters
lr = 0.0001
batch_size = 4
num_epochs = 10
lambda_cycle = 10
lambda_identity = 5
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)),
])
```

```python
dataset = ImageDataset("monet2photo", transform=transform)
```

```python
trainer = CycleGAN_Trainer(lr, batch_size, num_epochs, lambda_cycle, lambda_identity, train_dataset=dataset, device=device)
```

```python
trainer.train()
```
